```python
#В эти списки будем записывать полученные данные
price_cards = []
hp_cards = []
dspl_cards= []
ft_cards = []
id_cards = []
model_cards = []
year_cards = []
body_cards = []
color_cards = []
tr_cards = []
dr_cards = []
wh_cards = []
st_cards = []
own_cards = []
pts_cards = []
km_cards = []


# Перебираем ссылки
i=0
for car_url in car_links:
    req = requests.get(car_url, headers = HEADERS)
    req.encoding = 'utf-8'
    i+=1
    model = BeautifulSoup(req.text, 'html.parser')
    
    #номер объявления
    try:
        car_id = model.find('div', class_='CardHead__shortInfo_used').find_all(
            'div')[2].text.split(' ')[1]
        id_cards.append(car_id)
    except:
        id_cards.append(None)
        
    #марка, модель
    try:
        car_model = model.find('div', class_='CardHead').find('h1').text
        model_cards.append(car_model)
    except:
        model_cards.append(None)
        
    #год выпуска
    try:
        year = model.find('div', class_='CardOfferBody__leftColumn').find('a').text
        year_cards.append(year)
    except:
        year_cards.append(None)
        
    #пробег
    try:
        km = model.find('li', class_='CardInfoRow_kmAge').text.replace('\xa0', '')
        km_cards.append(km)
    except:
        km_cards.append(None)
        
    #тип кузова
    try:
        body = model.find('li', class_='CardInfoRow_bodytype').find('a').text
        body_cards.append(body)
    except:
        body_cards.append(None)
        
    #цвет
    try:
        color = model.find('li', class_='CardInfoRow_color').find('a').text
        color_cards.append(color)
    except:
        color_cards.append(None)
        
    #трансмиссия
    try:
        transmission = model.find('li', class_='CardInfoRow_transmission').find_all(
            'span')[1].text
        tr_cards.append(transmission)
    except:
        tr_cards.append(None)
        
    #привод
    try:
        drivetrain = model.find('li', class_='CardInfoRow_drive').find_all(
            'span')[1].text
        dr_cards.append(drivetrain)
    except:
        dr_cards.append(None)
        
    #руль
    try:
        wheel = model.find('li', class_='CardInfoRow_wheel').find_all('span')[1].text
        wh_cards.append(wheel)
    except:
        wh_cards.append(None)
        
    #состояние
    try:
        state = model.find('li', class_='CardInfoRow_state').find_all('span')[1].text
        st_cards.append(state)
    except:
        st_cards.append(None)
        
    #количество владельцев
    try:
        owners = model.find('li', class_='CardInfoRow_ownersCount').find_all(
            'span')[1].text.replace('\xa0', '')
        own_cards.append(owners)
    except:
        own_cards.append(None)
        
    #ПТС
    try:
        pts = model.find('li', class_='CardInfoRow_pts').find_all('span')[1].text
        pts_cards.append(pts)
    except:
        pts_cards.append(None)
       
    #Цена 
    try:
        price = model.find('div', class_='PriceUsedOffer').find(
            'span', class_='OfferPriceCaption__price').text.replace('\xa0', '')
        price_cards.append(price)
    except:
        price_cards.append(None)
        
    #мощность
    try:
        hp = model.find('li', class_='CardInfoRow_engine').find(
            'div').text.split('/')[1].replace('\xa0', '')
        hp_cards.append(hp)
    except:
        hp_cards.append(None)
        
    #объём двигателя
    try:
        dspl = model.find('li', class_='CardInfoRow_engine').find('div').text.split('/')[0]
        dspl_cards.append(dspl)
    except:
        dspl_cards.append(None)
            
    #тип топлива
    try:
        ft = model.find('li', class_='CardInfoRow_engine').find('div').text.split('/')[2]
        ft_cards.append(ft)
    except:
        ft_cards.append(None)
    
        
    print('читаем карточку:', i)
    time.sleep(0.1)
```

```python
#Складываем данные в датафрейм
data = pd.DataFrame()
data['price'] = price_cards
data['car_id'] = id_cards
data['model'] = model_cards
data['year'] = year_cards
data['km'] = km_cards
data['body'] = body_cards
data['color'] = color_cards
data['transmission'] = tr_cards
data['drivetrain'] = dr_cards
data['wheel'] = wh_cards
data['status'] = st_cards
data['owners'] = own_cards
data['pts'] = pts_cards
data['power'] = hp_cards
data['displacement'] = dspl_cards
data['fuel_type'] = ft_cards
data
```
